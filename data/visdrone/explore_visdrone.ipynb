```python
from pathlib import Path
import cv2, matplotlib.pyplot as plt, random

root = Path("images/train/images")     # iç içe klasöre işaret ediyor
images = list(root.glob("*.jpg"))
print("Görüntü sayısı:", len(images))

plt.figure(figsize=(10,10))
for i, img_path in enumerate(random.sample(images, min(4, len(images))), 1):
    img = cv2.cvtColor(cv2.imread(str(img_path)), cv2.COLOR_BGR2RGB)
    plt.subplot(2,2,i); plt.imshow(img); plt.axis("off"); plt.title(img_path.name)
plt.show()

```

```python
from pathlib import Path
import seaborn as sns, matplotlib.pyplot as plt

label_paths = Path("labels/train/labels").glob("*.txt")   # etiketlerde de çift klasör
classes = []
for lp in label_paths:
    for line in open(lp):
        classes.append(int(line.split()[0]))

sns.countplot(x=classes)
plt.xticks(range(10))
plt.title("Sınıf Frekansı")
plt.show()
```
